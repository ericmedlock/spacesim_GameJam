## Define data classes

### Subtask:
Create Python data classes for `CrewMember`, `Player`, and `Sector` to structure the game's data.


**Reasoning**:
Import the necessary decorator and define the data classes as specified in the instructions.



```python
from dataclasses import dataclass

@dataclass
class CrewMember:
    name: str
    role: str
    specialty: str
    stat: int

@dataclass
class Player:
    name: str
    rank: str
    role: str
    stats: int
    crew: list[CrewMember]

@dataclass
class Sector:
    name: str
    challenge: str
    difficulty: int
    risk: str
    reward: str
```

## Generate game data

### Subtask:
Implement functions to randomly generate crew members, a player, and space sectors with specified attributes and random values.


**Reasoning**:
Implement the functions to generate crew members, a player, and sectors as described in the instructions, using random values and the preloaded data. Then, generate the required number of each.



```python
import random

SECTOR_TYPES = [
    "Derelict Freighter",
    "Binary Star Anomaly",
    "Fungal Biofield",
    "Uncharted Asteroid Cluster",
    "Ghost Signal Beacon",
    "Wormhole Proximity Zone"
]

CHALLENGES = {
    "Derelict Freighter": "Board and investigate derelict systems",
    "Binary Star Anomaly": "Stabilize navigation through radiation waves",
    "Fungal Biofield": "Scan and analyze aggressive biospores",
    "Uncharted Asteroid Cluster": "Navigate dense and unpredictable asteroid fields",
    "Ghost Signal Beacon": "Investigate strange energy signatures",
    "Wormhole Proximity Zone": "Gather data near an unstable wormhole"
}

CREW_NAMES = [
    "Anya Sharma",
    "Jax Rourke",
    "Lena Petrova",
    "Kenji Tanaka",
    "Zara Khan",
    "Ben Carter",
    "Chloe Dubois"
]

CREW_ROLES = [
    "Engineer",
    "Scientist",
    "Tactician",
    "Diplomat",
    "Medic",
    "Pilot"
]

SPECIALTIES = ['Engineering', 'Science', 'Tactics', 'Diplomacy']

RISKS = [
    "Hull damage",
    "Crew injury",
    "System malfunction",
    "Resource loss",
    "Temporal displacement"
]

REWARDS = [
    "Valuable resources",
    "New technology schematic",
    "Navigational data",
    "Alliance opportunity",
    "Increased reputation"
]


def generate_crew_member():
    """Generates a random CrewMember object."""
    name = random.choice(CREW_NAMES)
    role = random.choice(CREW_ROLES)
    specialty = random.choice(SPECIALTIES)
    stat = random.randint(4, 9)
    return CrewMember(name=name, role=role, specialty=specialty, stat=stat)


def generate_player(crew_list):
    """Generates a Player object with fixed attributes and a list of crew."""
    name = random.choice(CREW_NAMES) # Player name can also be random
    rank = "Commander"
    role = "Captain"
    stats = random.randint(6, 10)
    return Player(name=name, rank=rank, role=role, stats=stats, crew=crew_list)


def generate_sector():
    """Generates a random Sector object."""
    sector_name = random.choice(SECTOR_TYPES)
    challenge = CHALLENGES[sector_name]
    difficulty = random.randint(3, 9)
    risk = random.choice(RISKS)
    reward = random.choice(REWARDS)
    return Sector(name=sector_name, challenge=challenge, difficulty=difficulty, risk=risk, reward=reward)

# Generate 3 crew members
crew_members = [generate_crew_member() for _ in range(3)]

# Generate a player
player = generate_player(crew_members)

# Generate 3 sectors
sectors = [generate_sector() for _ in range(3)]

# Print the generated objects to verify
print("Generated Crew Members:")
for crew in crew_members:
    print(crew)

print("\nGenerated Player:")
print(player)

print("\nGenerated Sectors:")
for sector in sectors:
    print(sector)
```

## Implement simulation logic

### Subtask:
Develop a simulation function that handles sector scanning, crew assignment, skill checks, and determines outcomes based on success or failure.


**Reasoning**:
Define the `simulate_sector_scan` function to handle the scanning process for a given sector and crew list.



```python
import random
import pandas as pd

def simulate_sector_scan(sector: Sector, crew_list: list[CrewMember]):
    """Simulates a sector scan with a randomly selected crew member."""
    assigned_crew = random.choice(crew_list)
    die_roll = random.randint(1, 6)
    skill_check = assigned_crew.stat + die_roll

    outcome = "Success" if skill_check >= sector.difficulty else "Failure"

    scan_results = {
        "Sector": sector.name,
        "Challenge": sector.challenge,
        "Assigned Crew": assigned_crew.name,
        "Crew Stat": assigned_crew.stat,
        "Die Roll": die_roll,
        "Skill Check": skill_check,
        "Difficulty": sector.difficulty,
        "Outcome": outcome
    }

    if outcome == "Success":
        scan_results["Result Detail"] = sector.reward
    else:
        scan_results["Result Detail"] = sector.risk

    return scan_results

# Simulate scans for the generated sectors
scan_outcomes = []
for sector in sectors:
    scan_outcomes.append(simulate_sector_scan(sector, crew_members))

# Create a DataFrame from the scan outcomes
scan_df = pd.DataFrame(scan_outcomes)

# Print the DataFrame
display(scan_df)
```
